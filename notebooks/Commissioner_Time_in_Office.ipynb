# Commissioner Time in Office

How long do commissioners spend in office? How many changes are there between elections? Who has served the longest? 

```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
```

```python
commissioners = pd.read_csv('../data/commissioners.csv')
```

```python
commissioners['start_date'] = pd.to_datetime(commissioners['start_date'])
commissioners['end_date'] = pd.to_datetime(commissioners['end_date'])
```

## Commissioner changes since last election

```python
commissioners[commissioners['start_date'] > datetime(2021,1,3)]
```

```python
(commissioners['start_date'] > datetime(2021,1,3)).sum()
```

```python
commissioners[(commissioners['end_date'] > datetime(2021,1,3)) & (commissioners['end_date'] < datetime(2023,1,1))]
```

```python
((commissioners['end_date'] > datetime(2021,1,3)) & (commissioners['end_date'] < datetime(2023,1,1))).sum()
```

## Shortest and Longest Serving Commissioners

```python
commissioners['time_in_office'] = commissioners['end_date'] - commissioners['start_date']
```

```python
commissioners.sort_values(by='time_in_office')
```

```python
commissioners.groupby('time_in_office').size()
```
