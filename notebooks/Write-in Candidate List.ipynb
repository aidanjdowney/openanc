# Write-in Candidate List

For the 2022 elections, make a list of all of the write-in candidates and their sources.

```python
import pandas as pd
```

```python
candidates = pd.read_csv('../data/candidates.csv')
districts = pd.read_csv('../data/districts.csv')
cd = pd.merge(candidates, districts, on='smd_id', how='inner')
```

```python
write_ins = cd[cd.candidate_status == 'Write-In Candidate'].copy()
```

```python
write_ins['SMD'] = write_ins['smd_name']
write_ins['Candidate Name'] = write_ins['candidate_name']
write_ins['Source'] = write_ins['manual_source']
```

```python
write_ins.sort_values(by='smd_id')[['SMD', 'Candidate Name', 'Source']]
```
